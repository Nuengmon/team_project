```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from load_data import *
from user_func import *
```

```python
df_covid = load_covid(['2020.1.20','2021.11.20'])
df_covid = df_covid[1:].T # 인덱스에 일자가 오도록함
df_covid = df_covid[:-1] #마지막 기간 총합 부분은 제외함

df_covid
```

# 시도별 확진자 증가 그래프   

그런데 그래프 보면 경기 서울이 너무 커서 다른곳이 안보임   
시도별 총인구수로 짤라서 1000만 이상 / 500만 이상 / 300만 이상 으로 나눠서 표현하는 방법도 고려

```python
fig_covid = go.Figure(
    layout=go.Layout(
        title=go.layout.Title(text="시도별 확진자 증가 그래프")
    )
)

for col in df_covid.columns:
    fig_covid.add_trace(go.Scatter(x=df_covid.index, y=df_covid[col], name=col) )
fig_covid.show()
```

# 최근 9, 10, 11월 시도별 확진자수 평균

```python
# 9월 지역별 확진자 평균
df_covid_9 = df_covid['2021.9.1':'2021.9.30']
df_covid_9 = df_covid_9.agg(np.mean)

# 10월 지역별 확진자 평균
df_covid_10 = df_covid['2021.10.1':'2021.10.31']
df_covid_10 = df_covid_10.agg(np.mean)

# 11월 지역별 확진자 평균
df_covid_11 = df_covid['2021.11.1':'2021.11.20']
df_covid_11 = df_covid_11.agg(np.mean)

df_covid_mean_recent = pd.DataFrame({
    '9월': df_covid_9,
    '10월': df_covid_10,
    '11월': df_covid_11
})

df_covid_mean_recent
```

# 아래 시도별 월별 확진자 평균을 보면 9월 이후 10월에 잠시 소강상태였다가 위드코로나와 함께 11월 에 증가함

```python
fig_covid_mean = go.Figure(data=[
    go.Bar(x=df_covid_mean_recent.index, y=df_covid_mean_recent['9월'], name='9월 확진자 평균'),
    go.Bar(x=df_covid_mean_recent.index, y=df_covid_mean_recent['10월'], name='10월 확진자 평균'),
    go.Bar(x=df_covid_mean_recent.index, y=df_covid_mean_recent['11월'], name='11월 확진자 평균'),
])
fig_covid_mean.show()
```
